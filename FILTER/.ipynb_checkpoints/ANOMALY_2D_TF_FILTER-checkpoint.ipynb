```python
indir_data = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/RAW_ANOMALY/TCWV/'
outdir_TF = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/TF2D/TCWV/'
var = 'tcwv'
year = np.arange(1991,1992)
y = year

for y in year :
    _ds_m1 = xr.open_mfdataset(indir_data+'*'+var+'*'+str(y-1)+'*.nc', chunks = { 'latitude' : 1}, parallel=True)
    _ds_m1 = _ds_m1.isel(time = slice(-120*int(24/3),None))
    _ds = xr.open_mfdataset(indir_data+'*'+var+'*'+str(y)+'*.nc', chunks = { 'latitude' : 1}, parallel=True)
    _ds1 = xr.open_mfdataset(indir_data+'*'+var+'*'+str(y+1)+'*.nc', chunks = { 'latitude' : 1}, parallel=True)
    _ds1 = _ds1.isel(time = slice(None,120*int(24/3)))

    ds = xr.concat([_ds_m1,_ds,_ds1], dim='time', coords='minimal', compat='override')
    ds = ds.chunk({"time" : np.shape(ds.time.values)[0], "latitude": 1})

    tcwvhat = xrft.fft(ds['tcwv_ano'],
                    dim=['time','longitude'], true_phase=False, true_amplitude=True)

    tcwv = xrft.ifft(tcwvhat,
                    dim=['freq_time','freq_longitude'], true_phase=False, true_amplitude=True)
    
    tcwvhat = tcwvhat.real.to_dataset(name = 'TF_tcwv')
    tcwvhat['time'] = ds.time
#     tcwvhat.to_netcdf(outdir_TF + 'TF2D_tcwv_ERA5_3H_'+str(y)+'.nc', unlimited_dims={'time':True})
    
#     tcwv['time'] = ds.time
#     tcwv = tcwv.sel(time = slice(str(y)+"-01-01",str(y)+"-12-31"))
#     tcwv = tcwv.to_dataset(name = 'inverse_TF_tcwv')
# #     tcwv.to_netcdf(outdir_TF + 'TF2D_tcwv_ERA5_3H_'+str(y)+'.nc', unlimited_dims={'time':True})
    
    
    
```

```python
def Rossby_filter(varhat, periode, wav_min, wav_max):
    radius = 6371000
    
    freq_filter = 1 / periode
    wavelength_max = (np.pi*2*6371000/wav_min)/111.11
    wavelength_min = (np.pi*2*6371000/wav_max)/111.11
    
    varhat = xr.where(varhat.freq_time < -freq_filter,  0., varhat )
    varhat = xr.where(varhat.freq_time > freq_filter,  0., varhat )
    varhat = xr.where(varhat.freq_longitude < -wavelength_min,  0., varhat) 
    varhat = xr.where(varhat.freq_longitude > wavelength_max,  0., varhat) 

    return varhat
```
