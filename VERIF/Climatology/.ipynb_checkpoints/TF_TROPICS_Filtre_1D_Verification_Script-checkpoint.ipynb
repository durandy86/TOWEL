```python
import os, sys
import numpy as np

import xarray as xr
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import colors

plt.rc("figure", figsize=(16,8))
plt.rc("font", size=14)
```

```python
indir = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/RAW_CLIM/'
exp = 'clim_tcwv_brut_ERA5_1979_2020'
ds = xr.open_dataset(indir+exp+'.nc', chunks = {'dayofyear' : 1, 'hour' : 1}
                                 )
```

## Selection
Dans la cellule au-dessus on a sortie dans un dataset la climatologie brute de ERA5, et on souhaite "smooth" le signal. On ne garde que les premi√®res harmonique du signal.

```python
ds_Af_clim = ds.sel(latitude = 10, longitude = 0)
ds_Id_clim = ds.sel(latitude = -10, longitude = 80)
```

```python
ds_Af_clim['tcwv'].sel(hour = 18).plot()
ds_Af_clim['tcwv'].sel(hour = 6).plot()
```

```python
## On met les donn√©es dans un tableau 1D
signal_Id = ds_Id_clim['tcwv'].values
signal_Id_S = signal_Id.flatten()
signal_Af = ds_Af_clim['tcwv'].values
signal_Af_S = signal_Af.flatten()
```

```python
nbSampl = 4
nbHarm = 4
nbHarmRemove = nbSampl*nbHarm
```

```python
import xrft
```

```python
time_plot = pd.date_range("2000-01-01", freq="6H", periods=366 * 4)


da = xr.DataArray(
    data=signal_Id_S,
    dims=["time"],
    coords=dict(
        time=time_plot,
    ),
    attrs=dict(
        description="tcwv_clim",
        units="kg.m^-2",
    ),
)
da.plot()
```

```python
da_F = xrft.dft(da, true_phase=False, true_amplitude=True)
da_F.real.plot()
```

```python
ida_dft = xrft.idft(da_F, true_phase = False, true_amplitude=True) # Signal in direct space
ida_dft['time'] = time_plot
ida_dft.real.plot()
plt.plot(time_plot, signal_Id_S[:1464], linewidth=1, label='Original signal')
```

```python
da_F = xrft.dft(da, dim="time", true_phase=False, true_amplitude=True)
_da = xr.where(da_F.freq_time<= -5/(86400*366),  0., da_F )
_da = xr.where(_da.freq_time>= 5/(86400*366),  0., _da)

xx = _da.sel(freq_time = -3/(366*86400), method = 'nearest')
xx
```

```python
ida_dft = xrft.ifft(_da, true_phase=False, true_amplitude=True) # Signal in direct space
```

```python
ida_dft['time'] = time_plot
ida_dft.real.plot()

plt.plot(time_plot, signal_Id_S[:1464], linewidth=1, label='Original signal')
```
