```python
#!/usr/bin/env python
# coding: utf-8

import os, sys
os.environ["MALLOC_TRIM_THRESHOLD_"] = '0'
import numpy as np

import xarray as xr
import xarray.ufuncs as xu
import xrft
import pandas as pd

from matplotlib import pyplot as plt

import cartopy.crs as ccrs
import cartopy

plt.rc("figure", figsize=(12,10))
plt.rc("font", size=14)

from dask.distributed import Client, LocalCluster
#
# Initialisation d'un cluster de 32 coeurs
cluster = LocalCluster(processes=False, n_workers=1, threads_per_worker=4, protocol = 'tcp', host = '137.129.155.67')
client = Client(cluster)
client
```

```python
wave_OLR = ['OLR_TD', 'OLR_MJO', 'OLR_Kelvin', 'OLR_Rossby', 'OLR_MRG', 'OLR_EIG','OLR_WIG']
indir_data = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/FILTERED_ANOMALY/OLR/'
coeff = 1

ds = xr.open_mfdataset(indir_data + '*.nc', chunks = {'time' : 10})
ds = ds.assign_coords(lon = (((ds.lon + 180) % 360) - 180)).sortby('lon')
months =  [7,8,9] # for example
_ds = ds.sel(time = np.in1d(ds['time.month'], months), lon =slice(-5,5), lat = slice(9,19))
_ds = _ds.mean(dim = ['lat','lon'])

indir_RAW = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/RAW_ANOMALY/OLR/'
datasets = []
for y in range(2001,2019):
    ds_RAW = xr.open_mfdataset(indir_RAW + 'anom_OLR_brut_ERA5_3H_*' + str(y) + '.nc', chunks = {'time' : 10})
    datasets.append(ds_RAW)
    
ds_RAW = xr.concat(datasets, dim ='time')
ds_RAW = ds_RAW.assign_coords(lon = (((ds.lon + 180) % 360) - 180)).sortby('lon')
_ds_RAW = ds_RAW.sel(time = np.in1d(ds_RAW['time.month'], months), lon =slice(-5,5), lat = slice(9,19))
_ds_RAW = _ds_RAW.mean(dim = ['lat','lon'])    
    
indir_var = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/ANALYSIS/VARIANCE/ANOMALY_FILTERED/OLR/'
season = 'YEAR'
ds_var = xr.open_mfdataset(indir_var + 'OLR_JAS_box_A.nc')
ds_DDt_var = xr.open_mfdataset(indir_var + 'OLR_DDT_JAS_box_A.nc')
ds_std = ds_var**(0.5)
ds_DDt_var = ds_DDt_var**(0.5)

indir_var = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/ANALYSIS/VARIANCE/RAW_ANOMALY/'
ds_std_RAW = xr.open_mfdataset(indir_var + 'OLR_JAS_box_A.nc')
```

```python
ds_std['OLR_ano'] = ds_std_RAW['OLR_ano']
_ds['OLR_ano'] = _ds_RAW['OLR_ano']
```

```python
ds_norm = _ds/ds_std
ds_DDt = ds.differentiate('time', datetime_unit = "s")
# ds_DDt['OLR_ano']
```

```python
months =  [7,8,9] # for example
_ds_DDt = ds_DDt.sel(time = np.in1d(ds_DDt['time.month'], months), lon =slice(-5,5), lat = slice(9,19))
_ds_DDt = _ds_DDt.mean(dim = ['lat','lon'])
_ds_DDt = _ds_DDt/ds_DDt_var
```

```python
# ds_norm['OLR_ano_DT'] = _ds_DDt['OLR_ano']
for w in wave_OLR :
    ds_norm[w + '_DT'] = _ds_DDt[w]
    
ds_norm = ds_norm
```

```python
v1 = ds_norm['OLR_Kelvin'].compute()
v2 = ds_norm['OLR_Kelvin_DT'].compute()
```

```python
arctang = np.arctan2(v2,v1)
ampl = np.sqrt(v1**2 + v2**2)
vect_x = ampl*np.cos(arctang)
vect_y = ampl*np.sin(arctang)

arctang.sel(time=slice('2009-07-01','2009-07-10')).plot()
plt.show()

ampl.sel(time=slice('2009-07-01','2009-07-10')).plot()
plt.show()

vect_x.sel(time=slice('2009-07-01','2009-07-10')).plot()
plt.grid()
plt.show()

vect_y.sel(time=slice('2009-07-01','2009-07-10')).plot()
plt.grid()
plt.show()
```

```python
ds_norm['vectx'] = vect_x
ds_norm['vecty'] = vect_y
ds_norm
```

```python
ds_norm.sel(time=slice('2009-07-01','2009-07-10')).plot.scatter(x = 'vectx', y = 'vecty')
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.grid()
plt.show()
```
