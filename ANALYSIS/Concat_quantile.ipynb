```python
datasets = []
for i in range(105):
    ds = xr.open_mfdataset(outdir + '/TEST/lon_' + str(i) + '_quantile.nc' , chunks = {'latitude' : 1}, combine='by_coords', parallel=True)
    ds = ds.expand_dims('longitude')
    ds = ds.assign_coords(longitude = ds.longitude)
    datasets.append(ds.load())

ds = xr.concat(datasets, dim='longitude', coords='minimal', compat='override')
ds.to_netcdf(outdir + 'lat_0to15_lon_m26to26_quantile.nc' )
```
