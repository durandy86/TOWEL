# Quantile pour l'OLR et l'onde de Kelvin

```python
quantile = [0.01,0.1,0.25,0.5,0.75,0.9,0.99]
datasets = []
wave = 'Kelvin'
for q in quantile :
    ds = xr.open_mfdataset(indir_data_Quantile + wave + '/*' + str(int(q*100)) + '.nc', parallel = True)
    datasets.append(ds)
ds_quan = xr.concat(datasets , dim ='quantile')
ds

plt.rc("figure", figsize=(16,4))
cmap ='BrBG'
vmin , vmax = -40, 40
colorbar = {'label': 'OLR $W.m^{-2}$',
            'orientation': 'vertical' , 'ticks': np.arange(vmin,vmax+0.1,5),
            'extend' : 'max'}

#####################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.01).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap , vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 1% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q1.png')

#####################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.1).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap , vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 10% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q10.png')
#####################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.25).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 25% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q25.png')
#####################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.5).plot.contourf(ax=axis, levels = 101,
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 50% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q50.png')
#######################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.75).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 75% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q75.png')
##########################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.9).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 90% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q90.png')
##########################################################################################
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

ds_quan['OLR_' + wave].sel(quantile = 0.99).plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Quantile 99% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_Q99.png')

#############################################################################################
#############################################################################################
#############################################################################################
vmin , vmax = 0, 80
colorbar = {'label': 'OLR $W.m^{-2}$',
            'orientation': 'vertical' , 'ticks': np.arange(vmin,vmax+0.1,5),
            'extend' : 'max'}
diff_Q = ds_quan['OLR_' + wave].sel(quantile = 0.99) - ds_quan['OLR_' + wave].sel(quantile = 0.01)
fig, axis = plt.subplots(
    1, 1, subplot_kw={'projection': map_proj})

diff_Q.plot.contourf(ax=axis, levels = 101, 
                    transform=ccrs.PlateCarree(),
                    cmap=cmap, vmin = vmin, vmax = vmax,
                    cbar_kwargs=colorbar)
                    

axis.coastlines()
axis.gridlines(draw_labels=True)
gl = axis.gridlines(draw_labels=True)
gl.top_labels = False
gl.right_labels = False
plt.title('Ecart interquantile 99%-1% pour '+ wave)
plt.savefig('/cnrm/tropics/commun/DATACOMMUN/WAVE/RAPPORT/FIGURES/Propre/Rapport/Chapter3/QUANTILE/OLR/' 
            + wave + '_DIFF.png')
```
