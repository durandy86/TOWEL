```python
#!/usr/bin/env python
# coding: utf-8

import os, sys
import numpy as np

import xarray as xr
import pandas as pd

from dask.distributed import Client, LocalCluster
from dask.diagnostics import ProgressBar
#
# Initialisation d'un cluster de 32 coeurs
cluster = LocalCluster(processes=False, n_workers=1, threads_per_worker=8)
client = Client(cluster)
client
```

```python
#########################################################################################
def witchYear(y, _ds_clim, time):
    def isLeapYear(yearN):
        if ((yearN % 4 == 0) and (yearN % 100 != 0)) or (yearN % 400 == 0):
            reponse = True
        else:
            reponse = False
        return reponse
    ##################
    
    if isLeapYear(y) == False :
        _ds_clim = ds_clim.isel(time = slice(0,365*int(24/3)))
        _ds_clim['time'] = time
    else :
        _ds_clim = ds_clim
        _ds_clim['time'] = time
    
    return _ds_clim
```

```python
#################################################################################
indir_clim = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/SMOTHED_CLIM/'
indir_data = '/cnrm/tropics/commun/DATACOMMUN/ERA5/0.25/netcdf/pl_3h/'

var = ['vo']

outdir = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/RAW_ANOMALY/'
year = np.arange(1990,2021)
```

```python
for v in var :
    ds_clim = xr.open_dataset(indir_clim + 'clim_' + v + '_smooth_ERA5_3H_1990_2020.nc', chunks = {'time' : 1, 'level' : 1})
    ds_data = xr.open_mfdataset(indir_data+'*_' + v + '_*.nc', chunks = {'time' : 1, 'level' : 1}, parallel=True)
    ds_clim = ds_clim.sel(latitude = slice(40.01, -40.01))
    ds_data = ds_data.sel(latitude = slice(40.01, -40.01))

    outdir = '/cnrm/tropics/commun/DATACOMMUN/WAVE/NO_SAVE/DATA/RAW_ANOMALY/' + v + '/'
    for y in year :
        print(y)
        ds = ds_data.sel(time = str(y))
        _ds_clim = witchYear(y, ds_clim, ds['time'])
        da = ds[v] - _ds_clim[v + '_smooth']
        da = da.to_dataset(name = v + "_ano")
        da.to_netcdf(outdir + 'anom_' + v + '_brut_ERA5_3H_'+str(y)+'.nc', unlimited_dims={'time':True})
```
